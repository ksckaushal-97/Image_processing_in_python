```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage.color
import skimage.util
```

```python
img = cv2.imread('./data/20181129_090518.jpg')[...,::-1]
sum_ =img.sum(axis =2).reshape(3024,4032,1)
norm =img/sum_
RGB =norm*3
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Image')
plt.subplot(1, 2, 2)
plt.imshow(RGB[:,:,0]>1.5, cmap='gray', interpolation = 'None')
plt.title('Threshold');
```
